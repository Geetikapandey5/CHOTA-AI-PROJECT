```python
import pyttsx3
import speech_recognition as sr
import requests
from datetime import datetime
import webbrowser

engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()

def listen():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("CHOTA is listening...")
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            print("You said:", text)
            return text
        except sr.UnknownValueError:
            speak("I could not understand what was said.")
            return None
        except sr.RequestError:
            speak("Failed to connect.")
            return None

def get_weather(location):
    api_key = '9e6fbfe5f1e7ae0d66238c2d6556e2ac'  
    url = f"http://api.openweathermap.org/data/2.5/weather?q={location}&appid={api_key}&units=metric"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        temperature = data['main']['temp']
        description = data['weather'][0]['description']
        return f"The weather in {location} is {description} with a temperature of {temperature}Â°C."
    else:
        return "Weather data not found. Please check the location and try again."

def is_greeting(command):
    variations = ["hello chhota", "hello chota", "hello chote", "hello chotey", "hello chotu"]
    return any(var in command.lower() for var in variations)

def google_search(query):
    # Create a Google search URL with the query
    search_url = f"https://www.google.com/search?q={query}"
    # Open the default browser with the search URL
    webbrowser.open(search_url)
    speak(f"Here are the results for '{query}' on Google.")

def play_youtube(query):
    youtube_search_url = f"https://www.youtube.com/results?search_query={query}"
    # Open the default browser with the YouTube
    webbrowser.open(youtube_search_url)
    speak(f"Playing '{query}' on YouTube.")

def main():
    speak("Hello, I am CHOTA, your personal assistant. What can I do for you today?")

    while True:
        command = listen()

        if command is None:
            continue

        if command.lower() == 'stop':
            speak("Goodbye! Have a nice day. If you want any help, I am here for you.")
            break

        if "who are you" in command.lower():
            speak("I am CHOTA, your virtual assistant. How can I help you?")
            continue
        
        if "who am i" in command.lower():
            speak("You are my owner. How can I help you?")
            continue
        
        if is_greeting(command):
            speak("Hello! How can I help you?")
            continue
        
        if command.lower().startswith("google"):
            search_query = command.lower().split("google")[1].strip()
            if search_query:
                google_search(search_query)
            else:
                speak("Please specify what you want to Google.")
            continue
        
        if "play youtube" in command.lower():
            youtube_query = command.lower().split("play youtube")[1].strip()
            if youtube_query:
                play_youtube(youtube_query)
            else:
                speak("Please specify what you want to play on YouTube.")
            continue
        
        if "weather in" in command.lower() or "today's weather" in command.lower():
            city = command.lower().split("weather in")[-1].strip() or command.lower().split("today's weather")[-1].strip()
            if city:
                weather_info = get_weather(city)
                speak(weather_info)
            else:
                speak("Please specify a city to get the weather information.")
            continue
        
        if "what's the time" in command.lower() or "what is the time" in command.lower():
            current_time = datetime.now().strftime('%H:%M:%S')
            speak(f"The current time is {current_time}.")
            continue
        
if __name__ == "__main__":
    main()
```
